```python
from IPython.display import clear_output
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Agents():    

    x = 0
    y = 0
    d1 = 0
    steps = 0
    color = ''

    def step(self):
        self.steps += 1
        sign_x = (-1)**np.random.randint(0, 2) 
        sign_y = (-1)**np.random.randint(0, 2)
        rand_x = np.random.sample()*self.d1
        rand_y = np.random.sample()*np.sqrt(self.d1**2 - rand_x**2)
        self.x += sign_x*rand_x
        self.y += sign_y*rand_y

    def paint(self):
            return [self.x, self.y, self.color]


# здоровый (состояние 0)
class Healthy(Agents):

    def __init__(self, d1):
        super().__init__()
        self.color = 'green'
        self.x = np.random.sample()
        self.y = np.random.sample()
        self.d1 = d1


# инфицированный (состояние 1)
class Infected(Agents):

    d2 = 0    # заражает агента при приближении к нему не более чем на d2
    p1 = 0    # вероятность заразить
    td1 = 0   # за td1 шагов переходит в состояние 3 с вероятностью pc1
    pc1 = 0   # вероятность перейти в состояние 3. если не перешел, переходит в состояние 4

    def __init__(self, d1, d2, p1, td1, pc1):
        super().__init__()
        self.color = '#ff9999'
        self.x = np.random.sample()
        self.y = np.random.sample()
        self.d1 = d1
        self.d2 = d2
        self.p1 = p1
        self.td1 = td1
        self.pc1 = pc1

    def inf(self):
        probability = int(self.p1*100)
        percent_tr = np.hstack([np.zeros(100 - probability, dtype=np.int), np.ones(probability, dtype=np.int)])
        index_rand = np.random.randint(0, 100)
        return percent_tr[index_rand]

    def chance_three(self):
        probability = int(self.pc1*100)
        percent_tr = np.hstack([np.zeros(100 - probability, dtype=np.int), np.ones(probability, dtype=np.int)])
        index_rand = np.random.randint(0, 100)
        return percent_tr[index_rand]




# легко больной (состояние 2)
class EasilySick(Agents):

    d2 = 0  # заражает агента при приближении к нему не более чем на d2
    p2 = 0  # вероятность заразить

    def __init__(self, d1, d2, p2):
        super().__init__()
        self.color = 'red'
        self.x = np.random.sample()
        self.y = np.random.sample()
        self.d1 = d1
        self.d2 = d2
        self.p2 = p2

    def inf(self):
        probability = int(self.p2*100)
        percent_tr = np.hstack([np.zeros(100 - probability, dtype=np.int), np.ones(probability, dtype=np.int)])
        index_rand = np.random.randint(0, 100)
        return percent_tr[index_rand]


# тяжело больной (состояние 3)
class SeriouslyIll(Agents):

    d2 = 0    # заражает агента при приближении к нему не более чем на d2
    p3 = 0    # вероятность заразить
    td2 = 0   # если не умирает за td2 шагов, то переходит в состояние 4
    pd1 = 0   # вероятность умереть на каждом шаге

    def __init__(self, d2, p3, td2, pd1):
        super().__init__()
        self.color = '#990000'
        self.x = np.random.sample()
        self.y = np.random.sample()
        self.d1 = d2
        self.d2 = d2
        self.p3 = p3
        self.td2 = td2
        self.pd1 = pd1

    def inf(self):
        probability = int(self.p3*100)
        percent_tr = np.hstack([np.zeros(100 - probability, dtype=np.int), np.ones(probability, dtype=np.int)])
        index_rand = np.random.randint(0, 100)
        return percent_tr[index_rand]

    def die(self):
        probability = int(self.pd1*100)
        percent_tr = np.hstack([np.zeros(100 - probability, dtype=np.int), np.ones(probability, dtype=np.int)])
        index_rand = np.random.randint(0, 100)
        return percent_tr[index_rand]


# иммунный (состояние 4)    
class Immune(Agents):

    def __init__(self, d1):
        super().__init__()
        self.color = '#75c1ff'
        self.x = np.random.sample()
        self.y = np.random.sample()
        self.d1 = d1
```

# Параметры агентов

```python
# здоровый (состояние 0)
d1 = 0.3  # перемещается за один шаг не более чем на d1


# инфицированный (состояние 1) | двигается как здоровый
d2 = 0.1  # заражает агентов при приближении к ним не более чем на d2 с вероятностью p1
p1 = 0.2  # вероятность заразить
td1 = 3   # за td1 шагов переходит в состояние 3 с вероятностью pc1
pc1 = 0.4  # вероятность перейти в состояние 3, если не перешел, переходит в состояние 4


# легко больной (состояние 2) | двигается не более чем на d1/2 
p2 = 0.3  # заражает агентов при приближении к ним не более чем на d2 с вероятностью p2


# тяжело больной (состояние 3) | двигается не более чем на d2
p3 = 0.5  # заражает агентов при приближении к ним не более чем на d2 с вероятностью p3
pd1 = 0.2  # вероятность умереть на каждом шаге
td2 = 3  # если не умирает в течении td2 шагов, то переходит в состояние 4


# иммунный (состояние 4) | двигается не более чем на d1, не заражается, не заражает других
```

# Запускаем эпидемию

```python
health = [Healthy(d1) for _ in range(1000)]
infec = [Infected(d1, d2, p1, td1, pc1) for _ in range(10)]
easy_sick = [EasilySick(d1, d2, p2) for _ in range(1)]
serious = [SeriouslyIll(d2, p3, td2, pd1) for _ in range(1)]
immun = [Immune(d1) for _ in range(1)]


health_stat = {'move': [], 'stat': []}
infec_stat = {'move': [], 'stat': []}
easy_stat = {'move': [], 'stat': []}
serious_stat = {'move': [], 'stat': []}
immun_stat = {'move': [], 'stat': []}
dead_stat = {'move': [], 'stat': []}
n_deaths = 0


for move in range(10):

    _ = [i.step() for i in health]
    _ = [i.step() for i in infec]
    _ = [i.step() for i in easy_sick]
    _ = [i.step() for i in serious]
    _ = [i.step() for i in immun]

    xy_health = np.array([i.paint()[:2] for i in health])
    xy_infec = np.array([i.paint()[:2] for i in infec])
    xy_immun = np.array([i.paint()[:2] for i in immun])
    xy_serious = np.array([i.paint()[:2] for i in serious])
    xy_easy_sick = np.array([i.paint()[:2] for i in easy_sick])


    # графики

    health_stat['move'].append(move)
    health_stat['stat'].append(len(health))

    infec_stat['move'].append(move)
    infec_stat['stat'].append(len(infec))

    easy_stat['move'].append(move)
    easy_stat['stat'].append(len(easy_sick))

    serious_stat['move'].append(move)
    serious_stat['stat'].append(len(serious))

    immun_stat['move'].append(move)
    immun_stat['stat'].append(len(immun))

    dead_stat['move'].append(move)
    dead_stat['stat'].append(n_deaths)


    clear_output(True)
    plt.figure(figsize=(15,6))

    plt.subplot (1, 2, 2)
    plt.plot(health_stat['move'], health_stat['stat'], label = 'Здоровые', color = 'green')
    plt.plot(infec_stat['move'], infec_stat['stat'], label = 'Инфицированные', color = '#ff9999')
    plt.plot(easy_stat['move'], easy_stat['stat'], label = 'Легко больные', color = 'red')
    plt.plot(serious_stat['move'], serious_stat['stat'], label = 'Тяжело больные', color = '#990000')
    plt.plot(immun_stat['move'], immun_stat['stat'], label = 'Иммунные', color = '#75c1ff')
    plt.plot(dead_stat['move'], dead_stat['stat'], label = 'Смертность', color = '#000000')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=2)

    plt.subplot (1, 2, 1)
    if len(xy_health) != 0:
        plt.scatter(xy_health[:,0], xy_health[:,1], label = 'Здоровые', color = 'green')

    if len(xy_infec) != 0:
        plt.scatter(xy_infec[:,0], xy_infec[:,1], label = 'Инфицированные', color = '#ff9999')

    if len(xy_immun) != 0:
        plt.scatter(xy_immun[:,0], xy_immun[:,1], label = 'Иммунные', color = '#75c1ff')

    if len(xy_serious) != 0:
        plt.scatter(xy_serious[:,0], xy_serious[:,1], label = 'Тяжело больные', color = '#990000')

    if len(xy_easy_sick) != 0:
        plt.scatter(xy_easy_sick[:,0], xy_easy_sick[:,1], label = 'Легко больные', color = 'red')
        
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=2)
    
#     plt.savefig(f'{move}.png')
    plt.show()


    # infec

    neighbors = np.sqrt(np.sum((xy_health-xy_infec.reshape(xy_infec.shape[0], 1, 2))**2, 2)) <= d2
    index_neighbors = np.array(np.where(neighbors==True)).T

    index_contact = []
    for i, j in index_neighbors:
        if (infec[i].inf() == 1) and (j not in index_contact):
            new_infec = Infected(d1, d2, p1, td1, pc1)
            new_infec.x = health[j].paint()[0]
            new_infec.y = health[j].paint()[1]
            infec.append(new_infec)
            index_contact.append(j)

    contact = []
    for i in index_contact:
        contact.append(health[i])

    for i in contact:
        health.remove(i)


    infec_steps = []
    for i in infec:
        if i.steps == td1:
            if i.chance_three() == 1:  # переходит в состояние 3
                new_serious = SeriouslyIll(d2, p3, td2, pd1)
                new_serious.x = i.paint()[0]
                new_serious.y = i.paint()[1]
                serious.append(new_serious)

            else:  # переходит в состояние 4
                new_immun = Immune(d1)
                new_immun.x = i.paint()[0]
                new_immun.y = i.paint()[1]
                immun.append(new_immun)
            infec_steps.append(i)

    for i in infec_steps:
        infec.remove(i)

    #  -----



    # easy sick

    xy_health = np.array([i.paint()[:2] for i in health])

    neighbors = np.sqrt(np.sum((xy_health-xy_easy_sick.reshape(xy_easy_sick.shape[0], 1, 2))**2, 2)) <= d2
    index_neighbors = np.array(np.where(neighbors==True)).T

    index_contact = []
    for i,j in index_neighbors:
        if (easy_sick[i].inf() == 1) and (j not in index_contact):
            new_infec = Infected(d1, d2, p1, td1, pc1)
            new_infec.x = health[j].paint()[0]
            new_infec.y = health[j].paint()[1]
            infec.append(new_infec)
            index_contact.append(j)

    contact = []
    for i in index_contact:
        contact.append(health[i])

    for i in contact:
        health.remove(i)

    # -----



    # serious

    xy_health = np.array([i.paint()[:2] for i in health])
    xy_serious = np.array([i.paint()[:2] for i in serious])

    neighbors = np.sqrt(np.sum((xy_health-xy_serious.reshape(xy_serious.shape[0], 1, 2))**2, 2)) <= d2
    index_neighbors = np.array(np.where(neighbors==True)).T

    index_contact = []
    for i, j in index_neighbors:
        if (serious[i].inf() == 1) and (j not in index_contact):
            new_infec = Infected(d1, d2, p1, td1, pc1)
            new_infec.x = health[j].paint()[0]
            new_infec.y = health[j].paint()[1]
            infec.append(new_infec)
            index_contact.append(j)

    contact = []
    for i in index_contact:
        contact.append(health[i])

    for i in contact:
        health.remove(i)


    dead = []
    for i in serious:
        if i.die() == 1:
            dead.append(i)
            n_deaths += 1

    for i in dead:
        serious.remove(i)


    recovered = []
    for i in serious:
        if i.steps == td2:  # переходит в состояние 4
            new_immun = Immune(d1)
            new_immun.x = i.paint()[0]
            new_immun.y = i.paint()[1]
            immun.append(new_immun)
            recovered.append(i)

    for i in recovered:
        serious.remove(i)

    # -----


print('Healthy: ', len(health))
print('Infected: ', len(infec))
print('Easily sick: ', len(easy_sick))
print('Seriously ill: ', len(serious))
print('Immune: ', len(immun))
print('Number of deaths: ', n_deaths)
```
